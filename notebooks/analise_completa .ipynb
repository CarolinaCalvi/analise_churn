## Analise Exploratoria

Utiliza-se a matriz de correlação para entender quais variáveis podem ser analisadas a fim de se obter uma melhor visão do churn. Necessita-se transformar os dados categóricos em valores numéricos para que ocorra a correlação.

```python
'''Codigo para a geração da matriz de correlção'''

df2 = df1[['created_at','updated_at','deleted_at','status',
           'last_date_purchase','average_ticket','items_quantity',
           'all_revenue','all_orders','recency','marketing_source','churn','regiao'
        ]]

# Usando o one-hot encoding   
categorical_encoded = pd.get_dummies(df2[['status','marketing_source','regiao']], prefix='')
df2 = pd.concat([df2, categorical_encoded], axis=1)
df2 = df2.drop(['status','marketing_source','regiao'], axis=1)

corr_matrix = df2.corr()

# Filtrando a matrix de correlacao para mostrar apenas a variavel 'churn'
churn_corr = corr_matrix['churn'].drop('churn').sort_values(ascending=False)

# Configurando o tamanho da imagem
fig, ax = plt.subplots(figsize=(10, 4))

# Grafico de barras
sns.barplot(x=churn_corr.values, y=churn_corr.index, color='#9B4E96', ax=ax)

# Configurando titulos e mostrando o grafico
plt.title('Correlação focada no Churn',fontsize=12, fontweight='bold')
plt.xlabel('Correlação',fontweight='bold')
plt.ylabel('Variáveis',fontweight='bold')
plt.savefig('../graficos/correlacao_churn.png',dpi=300, bbox_inches='tight')
plt.show()
```

A partir deste gráfico baseado na matriz de correlação, podemos observar que as variáveis recency, status e deleted_at possuem um alto grau de correlação com o churn. Sendo assim, iremos analisar essas 3 variáveis prioritariamente.

#### Analisando a variavel recency x churn_real (clientes cancelados)

```python
'''Codigo para a geração de um histograma'''

# Query para identificar a quantidade de churn se baseando nos dias
query= '''
SELECT 
    recency,
    COUNT(deleted_at) AS churn 
    FROM df1
    WHERE deleted_at IS NOT NULL
    GROUP BY 1 
'''

churn_data = ps.sqldf(query)

# Criando uma lista de bins para setar corretamente o periodo que eu quero analisar (de mes em mes)
max_recency = 30 * ((max(churn_data['recency']) + 29) // 30)
bins = range(0, max_recency+31, 90)

# Histograma
plt.figure(figsize=(20,6))
plt.hist(churn_data['recency'],
         bins=bins,
         weights=churn_data['churn'],
         color='#A569BD')

# Adicionando valores no topo do grafico
for rect in plt.gca().patches:
    x = rect.get_x() + rect.get_width() / 2
    y = rect.get_height()
    plt.annotate(f'{y:.0f}',
                 (x, y),
                 xytext=(0, 1),
                 textcoords='offset points',
                 ha='center',
                 va='bottom',
                 fontsize=12,
                 fontweight='bold')

ax.grid(False)
plt.xlabel('Dias desde a ultima compra',fontweight='bold')
plt.ylabel('Quantidade de clientes',fontweight='bold')
plt.title('Churn a cada 90 dias', fontsize=16, fontweight='bold')
# Renomeando o eixo x
plt.xticks(range(0, max_recency+1, 90), [str(x) for x in range(0, max_recency+1, 90)])
plt.savefig('../graficos/churn_trimestral.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
'''Codigo para a geração de um histograma'''

# Criando uma lista de bins para setar corretamente o periodo que eu quero analisar (de ano em ano)
bins = range(0, max_recency+31, 365)

# Histograma
plt.figure(figsize=(10,5))
plt.hist(churn_data['recency'],
         bins=bins,
         weights=churn_data['churn'],
         color='#A569BD')
# Adicionando valores no topo do grafico
for rect in plt.gca().patches:
    x = rect.get_x() + rect.get_width() / 2
    y = rect.get_height()
    plt.annotate(f'{y:.0f}',
                 (x, y),
                 xytext=(0, 1),
                 textcoords='offset points',
                 ha='center',
                 va='bottom',fontsize=10,
                 fontweight='bold')

ax.grid(False)
plt.xlabel('Dias desde a última compra',fontweight='bold')
plt.ylabel('Quantidade de clientes',fontweight='bold')
plt.title('Churn ao longo dos anos', fontsize=16, fontweight='bold')
# Renomeando o eixo x
plt.xticks(range(0, max_recency+1, 365), [str(x) for x in range(0, max_recency+1, 365)])
plt.savefig('../graficos/churn_anual.png', dpi=300, bbox_inches='tight')
plt.show()
```

Com base nos gráficos, nota-se que a maior parte dos churn provém de clientes mais recentes. Percebe-se também que há churn de muitos anos ainda contando recency.

### Analisando o churn estimado


```python
df2 = df1.copy()
churn = []
for i, row in df2.iterrows():
    if(row['status'] == 'canceled'):
        churn.append(1)
    elif(row['status'] == 'paused'):
        churn.append(2)
    elif(row['all_orders'] == 0 and row['status'] not in ['canceled', 'paused']):
        churn.append(3)
    else:
        churn.append(0)
df2['churn_estimado'] = churn

df2['churn_estimado'].unique()
```

```python
'''Codigo para a geração de gráfico de barras'''

total=len(df2)
fig, ax = plt.subplots(figsize = (10,6))
g=sns.countplot(data = df2, x='churn_estimado',palette=['#BB8FCE', '#8E44AD','#8E44AD','#8E44AD'])
for p in g.patches:
    height = p.get_height()
    g.text(p.get_x()+p.get_width()/2.,
            height/2,
            f'{100*height/total:1.2f}%',
            ha="center", color='black',fontsize=12, fontweight='bold')
    
ax.set_xticklabels(['Não Churn', 'Churn Atual','Pausados','Sem Compras'],fontweight='bold')
plt.title('Porcentagem absoluta de Churn Estimado',fontsize=12, fontweight='bold')
plt.xlabel('')
plt.ylabel('Quantidade clientes',fontweight='bold')
plt.savefig('../graficos/churn_estimado_divido.png',dpi=300, bbox_inches='tight')
plt.show()
```

```python
'''Codigo para a geração de gráfico de barras'''

df2 = df1

query='''SELECT status, SUM(churn) AS churn
FROM (
    SELECT 
        (CASE WHEN status == 'active' THEN 'paused' ELSE status END) AS status,
        COUNT(status) AS churn
    FROM df1
    WHERE all_orders == 0 OR status == 'canceled' OR status =='paused'
    GROUP BY 1
)
GROUP BY 1
'''
df_query = ps.sqldf(query)
df_query['churn_estimado'] = df_query['churn'].cumsum()
df_query = df_query[['status','churn_estimado']]

# Creating bar plot
fig, ax = plt.subplots(figsize=(10, 4))
g = sns.barplot(data=df_query, x=['Churn atual', 'Churn estimado'],y='churn_estimado',color='#8E44AD',ax=ax)

for p in g.patches:
    height = p.get_height()
    g.text(p.get_x()+p.get_width()/2.,
            height/2,
            f'{100*height/total:1.2f}%',
            ha="center", color='black',fontsize=12, fontweight='bold')

# Formatting plot
ax.set_xticklabels(['Churn Atual','Churn Estimado'],fontweight='bold')
ax.set_xlabel('Churn',fontweight='bold')
ax.set_ylabel('Quantidade de clientes',fontweight='bold')
ax.set_title('Aumento do Churn (acumulado)',fontsize=12,fontweight='bold')
plt.savefig('../graficos/churn_estimado.png',dpi=300, bbox_inches='tight')
ax.set_xlabel(' ')
# Displaying plot
plt.show()
```

A partir deste gráfico podemos observar que o churn possui um grande potencial de crescimento em relação ao atual, indo de 5.05% para 22.41%.

## Próximos Passos

Resumo:

Durante a analise observamos dados relacionados ao churn que podem implicar em sua variacao. Por conta disso, foi observado que o churn provem dos clientes mais recentes em comparacao com todo o historico do churn (5 anos), dentre os canais de marketing que possuem investimento o whatsapp e telegram sao os que possuem um maior nivel de churn e que esse churn e diretamente proporcional a quantidade de clientes, sendo o Nordeste a regiao com mais assinantes do servico Petlove. 

Caso nenhuma ação seja tomada, é possível que o churn aumente de 5,05% para 14,76% devido às contas pausadas, representando um aumento de 271%. Ademais, há 765 "ativos" que não realizam compras, o que eleva ainda mais o potencial de aumento do churn, podendo ir para 22.41%. Portanto, é essencial que sejam implementadas medidas para minimizar os efeitos do churn e garantir a fidelização dos clientes.


### Sugestão 1: Adotar o NPS
Uma possível sugestão para minimizar o churn seria adotar o Net Promoter Score (NPS). Essa metodologia é utilizada para avaliar o grau de satisfação e fidelidade dos clientes em relação a determinada empresa ou produto. Por meio do NPS, é possível coletar feedbacks dos clientes, identificar deficiências no produto ou serviço e até mesmo encontrar as razões para a insatisfação dos clientes.

No caso da Petlove, o NPS pode ser uma forma de entender melhor as necessidades e desejos dos clientes, o que pode ajudar a desenvolver estratégias mais eficazes para manter a fidelidade dos mesmos, evitando o churn dos clientes de longa data. Além disso, é possível utilizar o NPS para converter clientes que cancelaram seus serviços, incentivando-os a retornar e tornar-se novamente assinantes.

### Sugestão 2: Oferecer incentivos
Outra sugestão para minimizar o churn seria oferecer incentivos nas primeiras semanas de assinatura, a fim de aumentar a fidelização dos clientes tendo em vista que grande parte dos clientes fiéis possuem compras recorrentes entre 18 e 54 dias. 

Uma estratégia interessante seria oferecer um cupom de desconto para a primeira compra, juntamente com uma oferta do serviço de assinatura, destacando as vantagens de se tornar um assinante. 

Outra estratégia seria a criação de superdescontos apenas para assinantes como, “Compre 1 leve 2” durante o período que os assinantes menos compram, aumentando então a frequência das compras e diminuindo o churn.

Além disso, a empresa pode desenvolver um programa de recompensas ou pontos para os clientes mais fiéis, que realizam compras frequentemente e mantêm-se ativos por um período prolongado. Isso pode incentivar os clientes a permanecerem na plataforma por mais tempo.

### Sugestão 3: Reativar clientes ativos sem compras
Há clientes ativos que não realizam compras na Petlove desde 2017 e que só aumentam ao longo dos anos. Para minimizar um possível churn, a empresa pode adotar estratégias para incentivá-los a realizar novas compras.

Uma possível solução seria entrar em contato com esses clientes por e-mail ou por mensagens via whatsapp e telegram, oferecendo cupons de desconto com prazo de validade limitado (com cerca de 1 mês de validade), para incentivar a realização de novas compras. Além disso, é importante que na mensagem esteja destacado as vantagens de se tornar um assinante.
